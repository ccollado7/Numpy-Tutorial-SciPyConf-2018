# Slicing de Array 1-dimensional

```python
a = [10,11,12,13,14] #Defino un array 1-dimensional
```

```python
a
```

```python
a[1:3] #Defino un "slicing" de la forma [Inicio incluido:Fin sin incluir:step]

#La cantidad de elementos que tendra el array es igual a Fin sin incluir-Inicio incluido. En este caso 3-1 = 2 elementos
```

```python
a[1:-2] # [Inicio incluido=Indice 1,Fin sin incluir=-2]

#Los indices positivos siempre comienzan con 0
#Los indices negativos siempren comienzan con -1
```

```python
a[-4:3] #Otra forma de ver el slicing del array. NO es la forma recomendable de realizar el slicing para un array ya que puede generar confusion
```

```python
a[:3] #En este caso estoy omitiendo el Inicio del slice y por lo tanto se considera que es el indice 0
```

```python
a[0:3] #Verifico el caso anterior verificando la inclusion del indice 0
```

```python
a[3:] #En este caso omito el Fin del slice, por lo tanto se consideran todos los elementos hasta el final
```

```python
a[-2:] #Esto es equivalente a "seleccionar los ultimos 2 elementos del array"
```

```python
a[::2] #En este caso tomo el array completo cada 2 elementos como "step", quedando [::step=2]
```

# Slicing de Array 2-dimensional

```python
#Creo un array 2D de 6x6 colocando numeros consecutivos del 0 al 35

#Observar que en la creacion del array utilizando np.arange no se incluye el numero 36 que se indica.36 elementos tendra el array desde el 0 al 35 incluido
a = np.arange(36).reshape(6,6)
```

```python
a
```

```python
a.ndim #Numero de dimensiones
```

```python
a.shape #Tamaño del array
```

```python
a.size #Cantidad de elementos del array
```

Al trabajar con arrays 2D debo especificar un indice doble ya que ahora tengo 2 dimensiones:

1° dimension = N° de Filas

2° dimension = N° de Columnas

```python
#Voy a obtener un slice del array 2D:
    #Fila de indice 0
    #Columnas de indices 3 y 4
    
a[0,3:5]

#Como en todo slice en la definicion de 3:5 el numero 5 NO ESTA INCLUIDO
```

```python
#Voy a obtener un slice del array 2D:
    #Desde fila de indice 4 en adelante
    #Desde Columnas de indice 4 en adelante

a[4:,4:]
```

```python
#Voy a ver como se selecciona una columna entera
    #Todas las filas
    #Columna de indice 3

a[:,2]
```

```python
#Voy a ver como se selecciona:
    #Desde fila de indice 2 hasta el final con step=2
    #Todas las columnas con step=2

a[2::2,::2]
```

Ejercico Practico



```python
#Creo un array 5x5 con numeros consecutivos desde 0 al 24

b = np.arange(25).reshape(5,5)
```

```python
b
```

```python
#Elementos en ROJO
#Seleccion las columnas de indices 1 y 3:
    #Todas las filas
    #Desde columna 1 al final con step=2

b[:,1::2]
```

```python
#Elementos en AMARILLO
#Selecciono la ultima fila:
    #Fila de indice 4
    #Todas las columnas

b[4,:]
```

```python
#Elementos en AZUL
#Selecciono lo siguiente:
    #Fila de indice 1 y 3
    #Columna de indice 0 y 2

b[1::2,0:4:2]
```
